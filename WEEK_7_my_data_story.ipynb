```python
# Define the name of the file you want to read
file_name = 'New Text Document.txt'

try:
    # Open the file in read mode ('r').
    # The 'with' statement ensures the file is automatically closed after its block is exited.
    with open(file_name, 'r') as file:
        # Read all the content of the file and store it in a variable.
        file_content = file.read()

        # Print the content that was read from the file to the console/notebook output.
        print(file_content)

except FileNotFoundError:
    # This block executes if the file specified by 'file_name' does not exist.
    print(f"Error: The file '{file_name}' was not found. Please ensure it's in the correct directory.")
except Exception as e:
    # This block catches any other unexpected errors that might occur during file handling.
    print(f"An unexpected error occurred: {e}")
```

```python
# Define the name of the file
file_name = 'New Text Document.txt'

# The item to add
new_item = '- Candy'

try:
    # Open the file in append mode ('a').
    # This will add content to the end of the file without overwriting existing content.
    with open(file_name, 'a') as file:
        # Add a newline character before the new item to ensure it's on a new line.
        file.write('\n' + new_item)
    print(f"'{new_item}' has been added to '{file_name}'.")

except Exception as e:
    print(f"An error occurred while adding the item: {e}")
```
